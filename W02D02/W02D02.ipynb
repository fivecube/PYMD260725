```python
# List: append, extend

# append means adding at the end

list_of_numbers = [1, 2, 3, 4, 5]

for i in range(1, 11):
    if i in list_of_numbers:
        continue
    list_of_numbers.append(i)


print(list_of_numbers)


# extend as the name says means to extend the list with another list elements

x = [1, 2, 3]

y = ["apple", "mango"]

x.extend(y)

print(x)
```

```python
alphabet_count = {}

for i in range(97, 123):
    alphabet_count[chr(i)] = False

string1 = "the quick brown fox jumps over a lazy dog"

for char in string1:
    if char not in alphabet_count:
        continue
    alphabet_count[char] = True

print(alphabet_count)
```

```python
list_of_dict_values = list(alphabet_count.values())

print(list_of_dict_values)
all_true = True

for boolean in list_of_dict_values:
    if boolean is True:
        continue
    else:
        all_true = False

print(all_true)
```

```python
# Scope and namespaces
```

```python
x = 10

def fun1():
    x = 5
    x = x + 1
    print('fun1', x)

fun1()

print('global', x)
```

```python
x = 10

def fun1(y):
    x = 5
    x = x + y
    print('fun1', x)

fun1(10)

fun1(20)

print('global', x)
```

```python
x = 10

def fun1():
    global x
    x = x + 1
    print('fun1', x)

fun1()

print('global', x)
```
